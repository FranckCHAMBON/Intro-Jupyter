# Outils de visualisation pour la physique

On dispose d'un tableau de données dans un fichier,
on aimerait afficher une courbe.

* Notre fichier est `bille.csv`, dans le même répertoire.

```python
def lire_un_nombre(texte):
    return float(texte.replace(",", "."))

t_list = []
x_list = []
y_list = []

num = 0
with open('bille.csv', 'r') as f:
    for ligne in f.readlines():
        num  = num + 1
        if num > 2:
            trois_mots = list(ligne.split())
            t, x, y = map(lire_un_nombre, trois_mots)
            t_list.append(t)
            x_list.append(x)
            y_list.append(y)
print("liste de t", t_list[:5])
print("liste de x", x_list[:5])
print("liste de y", y_list[:5])
print("taille", len(t_list))
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.plot(t_list, y_list, "ro")
plt.xlabel("Temps")
plt.ylabel("Valeurs de $y$")
plt.title("Représentation graphique de $y(t)$)")
plt.show()
```

```python
def calcul_vitesses(x_list, y_list, t_list):
    v_x = []
    v_y = []
    for n in range(1, len(t_list)-1):
        v_x.append((x_list[n+1] - x_list[n-1]) / (t_list[n+1] - t_list[n-1]))
        v_y.append((y_list[n+1] - y_list[n-1]) / (t_list[n+1] - t_list[n-1]))
    t_list = t_list[1:-1]
    return v_x, v_y, t_list
```

```python
v_x, v_y, t_2 = calcul_vitesses(x_list, y_list, t_list)
plt.plot(t_2, v_y, "ro")
plt.xlabel("Temps")
plt.ylabel("Valeurs de $y'$")
plt.title("Représentation graphique de $y'(t)$)")
plt.show()
```
