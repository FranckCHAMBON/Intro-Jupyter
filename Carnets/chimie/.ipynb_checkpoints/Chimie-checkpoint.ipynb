# Visualisation de molécules chimiques
> Le module `ase` doit être installé ; par exemple avec le paquet debian `python3-ase`.

```python
from IPython.display import HTML, display
from ipywidgets import interact
from  ase.build import  molecule
from ase.collections import g2
```

La liste des molécules disponibles :

```python
", ".join(g2.names)
```

```python
def dessine_molecule_en3D(nom):
    atomes = molecule(nom) 
    from tempfile import NamedTemporaryFile
    with NamedTemporaryFile('r+', suffix='.html') as ntf:
        atomes.write(ntf.name, format='html')
        ntf.seek(0)
        html = ntf.read()
    return HTML(html)
```

## Exemples

+ 1 seule molécule à dessiner

```python
dessine_molecule_en3D(nom="C2H6CHOH")
```

+ Un menu déroulant des molécules du catalogue

```python
interact(dessine_molecule_en3D, nom=g2.names);
```
