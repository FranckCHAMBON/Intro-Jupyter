# Visualisation de molécules chimiques
> Le module `ase` doit être installé ; par exemple avec le paquet debian `python3-ase`.

```python
from IPython.display import HTML
from  ase.build import  molecule
from ase.collections import g2
", ".join(g2.names)
```

```python
def dessine_molecule_en3D(nom_molecule):
    atomes = molecule(nom_molecule) 
    from tempfile import NamedTemporaryFile
    with NamedTemporaryFile('r+', suffix='.html') as ntf:
        atomes.write(ntf.name, format='html')
        ntf.seek(0)
        html = ntf.read()
    return HTML(html)
```

## Exemple

```python
dessine_molecule_en3D("C2H6CHOH")
```
