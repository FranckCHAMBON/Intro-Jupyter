```python
n = 100   # nb répétition de l'épreuve de Bernoulli
p=0.2    # probabilité de succés à l'épreuve de Bernoulli
EXP = 10000  # Taille de l'échantillon statistique


q=1-p  # proba d'échec
R = [] # stockage des résultats
for i in range(EXP):
    nbS = 0 # nombre de succés
    for exp in range(n):
        x = randint(1, 1000)
        if x < 1000*p:
            nbS += 1 # un succès de plus
    R.append( nbS ) # on ajoute nbS à notre liste R


# Graphique
## histograme -> résulats statistiques !
## scatter_plot -> courbe théorique !
lo, hi = 6, 34
histogram(R,range=[lo,hi], bins=hi-lo+1) + scatter_plot([[k,EXP * binomial(n, k) * p**k * q**(n-k)] for k in range(lo, hi+1)])
```
