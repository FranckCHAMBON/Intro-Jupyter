Jupyter est une application qui propose un carnet de notes (_notebook_) mêlant :
* zones de textes mis en forme, et enrichies
    + d'images,
    + de tableaux,
    + de liens,
* et des zones de code,
    + principament pour Python,
    + mais aussi bash, html, latex, ...
    
L'utilisation de Jupyter se fait :
* soit en test en ligne avec [Try Jupyter](https://jupyter.org/try) (sans inscription) ;
* soit en l'installant (conseil : via [Anaconda](https://docs.anaconda.com/anaconda/install/)) ;
* soit en ligne avec le service complet [CoCalc](cocalc.com) (inscription nécessaire) ;
* la lecture d'un _notebook_ est possible directement dans GitHub (avec une qualité moindre) ;
* la lecture peut se faire aussi avec ...

On écrit dans une cellule que l'on choisira :
* <img src="https://raw.githubusercontent.com/dcurtis/markdown-mark/master/png/208x128.png" height="30"> en `Markdown` pour le texte,
* <img src="https://www.python.org/static/community_logos/python-logo-master-v3-TM.png" height="30"> ou bien `Code` pour les parties en Python.

Une fois le contenu écrit, on l'exécute avec <kbd>MAJ.</kbd>+<kbd>ENTRÉE</kbd>
(ou bien <kbd>SHIFT</kbd>+<kbd>RETURN</kbd>).

## A] Les zones de texte
Jupyter propose ses zones de texte en [Markdown](https://fr.wikipedia.org/wiki/Markdown) ; ce qui permet une saisie de texte rapide avec une qualité de
rendu HTML. En fait, on peut écrire aussi quelques balises HTML lorsque les balises Markdown sont limitées.
> En particulier, pour insérer des images et choisir des attributs, on utilisera la balise html `<img>`.

Voici une liste de tutoriels pour écrire en `Markdown` :
* https://www.markdowntutorial.com/ _in english by GitHub_
* [Syntaxe assez complète](https://michelf.ca/projets/php-markdown/syntaxe/)
* [another, in english, by GitHub](http://agea.github.io/tutorial.md/)  <===== Il faudrait le traduire !!!!
* [Un résumé assez succinct](https://github.com/luong-komorebi/Markdown-Tutorial/blob/master/README_fr.md)
* [Élaboration et conversion de documents avec Markdown et Pandoc](https://enacit1.epfl.ch/markdown-pandoc/) ; pour les utilisateurs avancés.

Et voici un condensé des possibilités ! 


### L'emphase faible et forte
```
Du _texte en italique_ entouré de la balise tiret bas *ou* une étoile.
```
> Du _texte en italique_ entouré de la balise tiret bas *ou* une étoile.

```
Du **texte en gras** entouré de la balise double étoile, __ou__ double tiret.
```
> Du **texte en gras** entouré de la balise double étoile, __ou__ double tiret.

```
**_Un_** exemple _avec_ tous *les* cas *__possibles__*, **fort1** ou __fort2__.
```
> **_Un_** exemple _avec_ tous *les* cas *__possibles__*, **fort1** ou __fort2__.

```
**Un dernier _exemple_** : _une section italique avec un mot en **gras**._
```
> **Un dernier _exemple_** : _une section italique avec un mot en **gras**._

>> Il faut juste respecter les balises.

### Les entêtes
Pour créer l'architecture d'un document, on place des titres de niveaux variés.

```
# Gros titre
## Partie 1
### Section a
### Section b
#### sous section
Un paragraphe...
#### autre sous section
Un autre paragraphe
## Partie 2
### Section a
On peut avoir jusqu'à 6 niveaux de titres
```
